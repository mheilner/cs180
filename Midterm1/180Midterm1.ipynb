Stellar Property Correlation Values

```python
# Temp v Rad
plt.figure()
sns.scatterplot(data=star, x='temperature', y='radius', hue='star_type')
plt.xscale('log')
plt.yscale('log')
plt.show()
# Lum v Rad
sns.scatterplot(data=star, x='luminosity', y='radius', hue='star_type')
plt.xscale('log')
plt.yscale('log')
plt.show()
# Absolute magnitude vs. Radius
plt.figure()
sns.scatterplot(data=star, x='absolute_magnitude', y='radius', hue='star_type')
plt.yscale('log')
plt.show()
# AB Mag v Lumin
plt.figure()
sns.scatterplot(data=star, x='absolute_magnitude', y='luminosity', hue='star_type')
plt.yscale('log')
plt.show()
# AB Mag v TEMP
plt.figure()
sns.scatterplot(data=star, x='absolute_magnitude', y='temperature', hue='star_type')
plt.yscale('log')
plt.show()
```

```python
# Question 18 - correlation coefficient
rRadvTemp = np.corrcoef(star['temperature'],star['radius'])
print(rRadvTemp)
print(rRadvTemp[0,1]**2)
print()
rRadvLum = np.corrcoef(star['luminosity'],star['radius'])
print(rRadvLum)
print(rRadvLum[0,1]**2)
print()
rAMvLum = np.corrcoef(star['absolute_magnitude'],star['luminosity'])
print(rAMvLum)
print(rAMvLum[0,1]**2)
print()
rAMvRad = np.corrcoef(star['absolute_magnitude'],star['radius'])
print(rAMvRad)
print(rAMvRad[0,1]**2)
print()
rAMvTemp = np.corrcoef(star['absolute_magnitude'],star['temperature'])
print(rAMvTemp)
print(rAMvTemp[0,1]**2)
print()
```
